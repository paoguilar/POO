```python
from datetime import datetime, timedelta

class Pasajero:
    def __init__(self, nombre, pasaporte, contacto):
        self.nombre = nombre
        self.pasaporte = pasaporte
        self.contacto = contacto

    def mostrar_info(self):
        print(f"Nombre: {self.nombre}, Pasaporte: {self.pasaporte}, Contacto: {self.contacto}")

class Vuelo:
    def __init__(self, destino, fecha_hora, num_vuelo, asientos_disponibles):
        self.destino = destino
        self.fecha_hora = fecha_hora
        self.num_vuelo = num_vuelo
        self.asientos_disponibles = asientos_disponibles
        self.pasajeros = []

    def reservar_asientos(self, pasajero, num_asientos):
        if self.asientos_disponibles >= num_asientos:
            self.pasajeros.append((pasajero, num_asientos))
            self.asientos_disponibles -= num_asientos
            return True
        return False

    def cancelar_reservas(self, pasajero, num_asientos):
        for p, asientos in self.pasajeros:
            if p == pasajero:
                if asientos >= num_asientos:
                    self.pasajeros.remove((p, asientos))
                    self.asientos_disponibles += num_asientos
                    return True
        return False

    def mostrar_info(self):
        print(f"Vuelo {self.num_vuelo} a {self.destino} el {self.fecha_hora}")
        print(f"Asientos disponibles: {self.asientos_disponibles}")
        print("Pasajeros:")
        for pasajero, asientos in self.pasajeros:
            print(f"- {pasajero.nombre} ({asientos} asientos)")

    def agregar_pasajero(self, pasajero):
      self.pasajeros.append(pasajero)

class Reserva:
    def __init__(self, pasajero, vuelo, num_asientos, codigo_reserva):
        self.pasajero = pasajero
        self.vuelo = vuelo
        self.num_asientos = num_asientos
        self.codigo_reserva = codigo_reserva

    def mostrar_info(self):
        print(f"Reserva {self.codigo_reserva} para {self.pasajero.nombre}")
        print(f"Vuelo: {self.vuelo.num_vuelo} a {self.vuelo.destino} el {self.vuelo.fecha_hora}")
        print(f"Asientos: {self.num_asientos}")

    def cancelar_reserva(self):
        if self.vuelo.cancelar_reservas(self.pasajero, self.num_asientos):
            return True
        return False

class EuroFly:
    def __init__(self, nombre):
        self.nombre = nombre
        self.destinos = []
        self.horarios_salida = ["05:00", "20:00"]
        self.dias_salida = ["Lunes", "Miércoles", "Viernes"]
        self.vuelos = []
        self.reservas = []

    def agregar_destino(self, destino):
        self.destinos.append(destino)

    def eliminar_destino(self, destino):
        if destino in self.destinos:
            self.destinos.remove(destino)

    def mostrar_destinos(self):
        print("Destinos disponibles:")
        for destino in self.destinos:
            print(f"- {destino}")

    def crear_vuelo(self, destino, fecha, hora, num_vuelo, asientos_disponibles):
        if destino in self.destinos and hora in self.horarios_salida and fecha.strftime("%A") in self.dias_salida:
            fecha_hora = datetime.strptime(f"{fecha.strftime('%Y-%m-%d')} {hora}", "%Y-%m-%d %H:%M")
            vuelo = Vuelo(destino, fecha_hora, num_vuelo, asientos_disponibles)
            self.vuelos.append(vuelo)
            return vuelo
        return None

    def buscar_vuelos(self, destino=None, fecha=None):
        resultados = []
        for vuelo in self.vuelos:
            if (destino is None or vuelo.destino == destino) and (fecha is None or vuelo.fecha_hora.date() == fecha.date()):
                resultados.append(vuelo)
        return resultados

    def cancelar_vuelo(self, vuelo):
        if vuelo in self.vuelos:
            self.vuelos.remove(vuelo)
            return True
        return False

    def crear_reserva(self, pasajero, vuelo, num_asientos, codigo_reserva):
        if vuelo.reservar_asientos(pasajero, num_asientos):
            reserva = Reserva(pasajero, vuelo, num_asientos, codigo_reserva)
            self.reservas.append(reserva)
            return reserva
        return None
# Ejemplo de uso
eurofly = EuroFly("EuroFly")
eurofly.agregar_destino("París")
eurofly.agregar_destino("Londres")
eurofly.agregar_destino("Berlín")

fecha_vuelo = datetime(2024, 12, 11)  # Miércoles
vuelo1 = eurofly.crear_vuelo("París", fecha_vuelo, "05:00", "EF123", 100)
vuelo2 = eurofly.crear_vuelo("Londres", fecha_vuelo, "20:00", "EF456", 150)

if vuelo1 and vuelo2:
    pasajero1 = Pasajero("Ana García", "12345678A", "ana@example.com")
    pasajero2 = Pasajero("Luis Pérez", "87654321B", "luis@example.com")

    reserva1 = eurofly.crear_reserva(pasajero1, vuelo1, 2, "R123")
    reserva2 = eurofly.crear_reserva(pasajero2, vuelo2, 1, "R456")

    if reserva1 and reserva2:
        print("Reservas creadas con éxito:")
        reserva1.mostrar_info()
        reserva2.mostrar_info()

        print("\nInformación del vuelo:")
        vuelo1.mostrar_info()
```
