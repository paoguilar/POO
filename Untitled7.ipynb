```python
!pip install -q streamlit
!npm install localtunnel
import streamlit as st

# Clase base Nota
class Nota:
    def __init__(self, titulo, contenido):
        self.titulo = titulo
        self.contenido = contenido

    def mostrar(self):
        return f"{self.titulo}: {self.contenido}"

    def modificar(self, nuevo_contenido):
        self.contenido = nuevo_contenido

# Clases derivadas
class NotaTexto(Nota):
    def formatear_texto(self):
        return f"***{self.contenido}***"

class NotaLista(Nota):
    def __init__(self, titulo, contenido):
        super().__init__(titulo, contenido.split(","))

    def mostrar(self):
        return f"{self.titulo}:\n" + "\n".join(f"- {item}" for item in self.contenido)

class NotaImagen(Nota):
    def __init__(self, titulo, ruta_imagen):
        super().__init__(titulo, ruta_imagen)

    def mostrar(self):
        return f"Imagen: {self.titulo}"

# Clase BlocDeNotas
class BlocDeNotas:
    def __init__(self):
        self.notas = []

    def agregar_nota(self, nota):
        self.notas.append(nota)

    def mostrar_notas(self):
        return [nota.mostrar() for nota in self.notas]

    def buscar_nota(self, titulo):
        return next((nota for nota in self.notas if nota.titulo == titulo), None)

    def eliminar_nota(self, titulo):
        self.notas = [nota for nota in self.notas if nota.titulo != titulo]

# Interfaz con Streamlit
bloc = BlocDeNotas()
st.title("Bloc de Notas Interactivo")

tipo_nota = st.selectbox("Selecciona el tipo de nota", ["Texto", "Lista", "Imagen"])
titulo = st.text_input("Título de la nota")
contenido = st.text_input("Contenido de la nota (para listas, separar con comas)")

if st.button("Agregar Nota"):
    if titulo and contenido:
        if tipo_nota == "Texto":
            bloc.agregar_nota(NotaTexto(titulo, contenido))
        elif tipo_nota == "Lista":
            bloc.agregar_nota(NotaLista(titulo, contenido))
        elif tipo_nota == "Imagen":
            bloc.agregar_nota(NotaImagen(titulo, contenido))
        st.success("Nota agregada correctamente")
    else:
        st.error("Por favor, ingresa un título y contenido")

st.subheader("Notas Guardadas")
for nota in bloc.mostrar_notas():
    st.write(nota)

buscar_titulo = st.text_input("Buscar nota por título")
if st.button("Buscar"):
    nota_encontrada = bloc.buscar_nota(buscar_titulo)
    if nota_encontrada:
        st.write(nota_encontrada.mostrar())
    else:
        st.error("Nota no encontrada")

eliminar_titulo = st.text_input("Eliminar nota por título")
if st.button("Eliminar"):
    bloc.eliminar_nota(eliminar_titulo)
    st.success("Nota eliminada")
```
